```python
def FCFS(cy,curr,stc):
  st=0
  for i in cy:
    st+=abs(i-curr)
    curr=i
  return st*stc,curr
```

```python
def SCAN(cy,curr,stc,f,l):
  cy_copy=cy.copy()
  to_append=[curr,f,l]
  cy_copy+=to_append
  cy_copy.sort()
  ind=cy_copy.index(curr)
  if((cy_copy[ind+1]-curr)>(curr-cy_copy[ind-1])):
    cy1=list(reversed(cy_copy[:ind]))
    cy2=cy_copy[ind+1:0]
  else:
    cy1=cy_copy[ind+1:]
    cy2=list(reversed(cy_copy[1:ind]))
  st1,curr1=FCFS(cy1,curr,stc)
  st2,curr2=FCFS(cy2,curr1,stc)
  seq=cy1+cy2
  st=st1+st2
  return st,seq
```

```python
def LOOK(cy,curr,stc):
  cy_copy=cy.copy()
  cy_copy.append(curr)
  cy_copy.sort()
  ind=cy_copy.index(curr)
  if((cy_copy[ind+1]-curr)>(curr-cy_copy[ind-1])):
    cy1=list(reversed(cy_copy[:ind]))
    cy2=cy_copy[ind+1:]
  else:
    cy1=cy_copy[ind+1:]
    cy2=list(reversed(cy_copy[:ind]))
  st1,curr1=FCFS(cy1,curr,stc)
  st2,curr2=FCFS(cy2,curr1,stc)
  seq=cy1+cy2
  st=st1+st2
  return st,seq
```

```python
def C_SCAN(cy,curr,stc,f,l):
  cy_copy=cy.copy()
  to_append=[curr,f,l]
  cy_copy+=to_append
  cy_copy.sort()
  ind=cy_copy.index(curr)
  if((cy_copy[ind+1]-curr)>(curr-cy_copy[ind-1])):
    cy1=cy_copy[:ind]
    cy2=cy_copy[ind+1:]
  else:
    cy1=cy_copy[ind+1:]
    cy2=cy_copy[:ind]
  st1,curr1=FCFS(cy1,curr,stc)
  st2,curr2=FCFS(cy2,cy2[0],stc)
  seq=cy1+cy2
  st=st1+st2+(cy1[-1]-cy2[0])*stc
  return st,seq
```

```python
import matplotlib.pyplot as plt
def plot(seq,curr,title):
  X=[curr]+seq
  plt.plot(X,range(len(X)))
  plt.gca().invert_yaxis()
  plt.title(title)
  plt.tight_layout()
  plt.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
```

```python
cy=[10,22,20,2,40,6,38]
curr=20
stc=6
#assuming first position = 0 and last position = 50
f=0
l=50

tst,c=FCFS(cy,curr,stc)
print("FCFS")
print("Sequence: ",cy)
print("Total seek time: ",tst)
plt.subplot(2, 2, 1)
plot(cy,curr,'FCFS')

tst,seq=SCAN(cy,curr,stc,f,l)
print("\nSCAN")
print("Sequence: ",seq)
print("Total seek time: ",tst)
plt.subplot(2, 2, 2)
plot(seq,curr,'SCAN')

tst,seq=LOOK(cy,curr,stc)
print("\nLOOK")
print("Sequence: ",seq)
print("Total seek time: ",tst)
plt.subplot(2, 2, 3)
plot(seq,curr,'LOOK')

tst,seq=C_SCAN(cy,curr,stc,f,l)
print("\nC-SCAN")
print("Sequence: ",seq)
print("Total seek time: ",tst)
plt.subplot(2, 2, 4)
plot(seq,curr,'C-SCAN')
```
